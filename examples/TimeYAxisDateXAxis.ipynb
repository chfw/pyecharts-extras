```python
import pyecharts.options as opts
from pyecharts_extras.date_time import TimeYAxisDateXAxis, DateAxisOpts, TimeAxisOpts
from datetime import datetime, date

data1 = [
   datetime(2020,9,1,3,11,11),
   datetime(2020,9,1,9,11,11),
   datetime(2020,9,2,4,11,11),
   datetime(2020,9,3,5,11,11),
   datetime(2020,9,4,6,11,11),
   datetime(2020,9,5,1,11,11),
   datetime(2020,9,6,2,11,11),
   datetime(2020,9,1,1,11,11),
   datetime(2020,9,2,2,11,11),
   datetime(2020,9,3,3,11,11),
   datetime(2020,9,4,4,11,11),
   datetime(2020,9,5,5,11,11),
   datetime(2020,9,6,6,11,11),

]

data2 = [
    datetime(2020,8,1,1,11,11),
    datetime(2020,9,2,3,19,11),
    datetime(2020,9,3,3,19,11),
    datetime(2020,9,4,3,19,11),
    datetime(2020,9,5,4,19,11),
    datetime(2020,8,6,5,19,11),
    datetime(2020,8,3,6,25,11),
    datetime(2020,8,4,4,19,11),
    datetime(2020,8,5,7,19,11),
    datetime(2020,8,6,3,19,11),
]

all_data = data1 + data2
all_dates = sorted(list(set([str(d.date()) for d in all_data])))
chart = (TimeYAxisDateXAxis(init_opts=opts.InitOpts(height="1000px"))
        .set_global_opts(
            xaxis_opts=DateAxisOpts(axis_data=all_dates),
            yaxis_opts=TimeAxisOpts(max_=),
            tooltip_opts=opts.TooltipOpts(is_show=False))    
        .add_data('A', data1, symbol_size=10)
        .add_data('B', data2, symbol_size=20))
chart.render_notebook()
```
